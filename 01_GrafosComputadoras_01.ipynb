%timeit es un "comando mágico" de IPython (usado en entornos como Jupyter Notebook) que mide de forma muy precisa cuánto tarda en ejecutarse una sola línea de código.

Aquí tienes un ejemplo clásico comparando la velocidad de una comprensión de listas contra un bucle for tradicional.

**%timeit para una sola línea**


Para medir el tiempo de una sola instrucción, usas %timeit seguido del código.

Imaginemos que queremos crear una lista con los cuadrados de los primeros 1,000 números usando una comprensión de listas.

````python
%timeit [i**2 for i in range(1000)]
````

**%%timeit para múltiples líneas**

Si quieres medir el tiempo de un bloque de código de varias líneas, usas %%timeit (con dos %) al inicio de la celda.

Ahora mediremos la misma operación, pero usando un bucle for.


````python
%%timeit
cuadrados = []
for i in range(1000):
  cuadrados.append(i**2)
````

## 1.3 NetworkX

`networkx` es una librería de Python para crear, manipular y estudiar grafos y redes complejas. Soporta grafos simples, dirigidos, multigrafos y permite calcular medidas de análisis de redes.

**¿Para qué sirve?**

- Representar relaciones y estructuras en forma de grafos.
- Calcular propiedades como grados, caminos más cortos, centralidades.
- Visualizar redes de tamaño pequeño o mediano.

**¿Cómo instalarlo?**
```bash
pip install networkx
# o
conda install -c conda-forge networkx


````python

import networkx as nx

# Crear un grafo no dirigido
G = nx.Graph()

# Agregar nodos y aristas
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Calcular propiedades
grados = dict(G.degree())
camino_corto = nx.shortest_path(G, source=1, target=4)
clustering = nx.average_clustering(G)

print("Grados:", grados)
print("Camino más corto entre 1 y 4:", camino_corto)
print("Clustering promedio:", clustering)
````

````python
# Ejemplo 1
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [0, 1, 2, 3, 4]
y = [v**2 for v in x]

# Crear gráfico
plt.figure()
plt.plot(x, y, label="y = x^2", marker="o")
plt.title("Ejemplo de gráfico con Matplotlib")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Ejemplo 2

import networkx as nx
import matplotlib.pyplot as plt # 1. Importar la librería de visualización

# Crear un grafo no dirigido
G = nx.Graph()

# Agregar nodos y aristas
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Calcular propiedades
grados = dict(G.degree())
camino_corto = nx.shortest_path(G, source=1, target=4)
clustering = nx.average_clustering(G)

print("Grados:", grados)
print("Camino más corto entre 1 y 4:", camino_corto)
print("Clustering promedio:", clustering)

# --- Visualización del grafo ---

# 2. Dibujar el grafo
nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, font_weight='bold')

# 3. Mostrar el grafo en una ventana
plt.show()
````

## 1.5 OSMnx

`osmnx` es una librería de Python que permite descargar, construir, analizar y visualizar grafos de calles y otros datos geoespaciales directamente desde **OpenStreetMap**.

**¿Para qué sirve?**

- Obtener redes viales reales como grafos.
- Calcular estadísticas urbanas: densidad, conectividad, longitudes de calles.
- Visualizar mapas y rutas.

**¿Cómo instalarlo?**
```bash
pip install osmnx
# o
conda install -c conda-forge osmnx
```

Es recomendable usar con conda-forge por las dependencias geoespaciales.


````python

# Ejemplo con OSMnx: descargar y mostrar una red vial
import osmnx as ox

lugar = "Guadalajara, Jalisco, Mexico"

# Descargar red de calles
G_calles = ox.graph_from_place(lugar, network_type="drive")

# Estadísticas básicas
stats = ox.basic_stats(G_calles)
print("Número de nodos:", stats['n'])
print("Número de aristas:", stats['m'])

# Visualizar
ox.plot_graph(G_calles)
````

## 1.6 Pandas

`pandas` es una librería de Python para el análisis y manipulación de datos, que proporciona estructuras eficientes como `DataFrame` y `Series`.

**¿Qué es un DataFrame?**  
Es una tabla de datos con etiquetas en filas y columnas, muy similar a una hoja de cálculo o una tabla SQL.

**¿Para qué sirve?**

- Cargar y guardar datos en múltiples formatos (`csv`, `excel`, `json`, `sql`).
- Filtrar, seleccionar y transformar datos.
- Realizar operaciones estadísticas y de agrupamiento.
- Integrarse con otras librerías como `matplotlib` y `seaborn` para visualización.

**¿Cómo instalarlo?**
```bash
pip install pandas
# o
conda install -c conda-forge pandas


````python

import pandas as pd

# Crear un DataFrame desde un diccionario
datos = {
    "Nombre": ["Ana", "Luis", "María", "Carlos"],
    "Edad": [23, 35, 29, 41],
    "Ciudad": ["Guadalajara", "CDMX", "Monterrey", "Tijuana"]
}

df = pd.DataFrame(datos)

# Mostrar el DataFrame
print(df)
display(df)

# Filtrar personas mayores de 30
mayores_30 = df[df["Edad"] > 30]
print("\nMayores de 30 años:\n", mayores_30)

# Calcular edad promedio
edad_prom = df["Edad"].mean()
print("\nEdad promedio:", edad_prom)
````

# Comandos básicos de Pandas

| Uso | Nombre de comando | Para qué sirve | Sintaxis | Ejemplo |
|-----|-------------------|----------------|----------|---------|
| Leer archivo CSV | `read_csv` | Cargar un archivo CSV en un DataFrame | `pd.read_csv("ruta.csv")` | `df = pd.read_csv("datos.csv")` |
| Guardar como CSV | `to_csv` | Guardar un DataFrame en un archivo CSV | `df.to_csv("ruta.csv", index=False)` | `df.to_csv("resultados.csv", index=False)` |
| Filtrar filas | Filtrado condicional | Seleccionar filas que cumplen una condición | `df[condicion]` | `df[df["Edad"] > 30]` |
| Insertar fila | `loc` con nuevo índice | Agregar una nueva fila al DataFrame | `df.loc[nuevo_indice] = valores` | `df.loc[4] = ["Ana", 28, "CDMX"]` |
| Modificar fila | `loc` | Cambiar todos los valores de una fila existente | `df.loc[indice] = nuevos_valores` | `df.loc[2] = ["Luis", 40, "GDL"]` |
| Eliminar fila | `drop` | Borrar una fila por índice | `df.drop(indice, inplace=True)` | `df.drop(1, inplace=True)` |
| Insertar columna | Asignación directa | Crear una nueva columna en el DataFrame | `df["NuevaCol"] = valores` | `df["Puntaje"] = [80, 90, 85]` |
| Modificar columna | Asignación directa | Cambiar todos los valores de una columna | `df["Columna"] = nuevos_valores` | `df["Edad"] = df["Edad"] + 1` |
| Eliminar columna | `drop` con `axis=1` | Borrar una columna | `df.drop("Columna", axis=1, inplace=True)` | `df.drop("Edad", axis=1, inplace=True)` |
| Modificar celda | `at` o `loc` | Cambiar el valor de una celda específica | `df.at[fila, columna] = valor` | `df.at[0, "Edad"] = 35` |
| Seleccionar columna | Indexado por nombre | Obtener los valores de una columna | `df["Columna"]` | `df["Nombre"]` |
| Seleccionar varias columnas | Lista de nombres | Obtener varias columnas a la vez | `df[["Col1", "Col2"]]` | `df[["Nombre", "Edad"]]` |
| Ordenar filas | `sort_values` | Ordenar por el valor de una o varias columnas | `df.sort_values("Columna")` | `df.sort_values("Edad", ascending=False)` |
| Renombrar columnas | `rename` | Cambiar el nombre de una o más columnas | `df.rename(columns={"Viejo":"Nuevo"})` | `df.rename(columns={"Edad":"Años"}, inplace=True)` |
| Describir datos | `describe` | Obtener estadísticas descriptivas | `df.describe()` | `df.describe()` |
| Contar valores | `value_counts` | Contar ocurrencias en una columna | `df["Columna"].value_counts()` | `df["Ciudad"].value_counts()` |
| Reemplazar valores | `replace` | Cambiar valores específicos en columnas o filas | `df.replace(valor_ant, valor_nuevo)` | `df["Ciudad"].replace("CDMX", "Ciudad de México", inplace=True)` |
